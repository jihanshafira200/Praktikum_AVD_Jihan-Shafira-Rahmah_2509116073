## **Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **Eksplorasi Data (EDA)**

## **Comparison(Perbandingan)**

```python
penjualan_country = df.groupby('Country')['Coffee_Intake'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=penjualan_country.index, y=penjualan_country.values, palette='viridis', hue=penjualan_country.index, legend=False)
plt.title('Total Coffee Intake Based on Country')
plt.xlabel('Country')
plt.ylabel('Total Coffee Intake')
plt.xticks(rotation=45)
plt.show()
```

Berdasarkan grafik dapat dilihat bahwa total konsumsi kopi berbeda di setiap negara beberapa negara seperti Canada, Norway, China dan UK memiliki total coffee intake yang relatif lebih tinggi dibandingkan negara lain, sementara negara seperti USA, Brazil dan Mexico berada pada tingkat yang lebih rendah.

## **Composition(Komposisi)**

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Pastel2.colors
)
plt.title('Proporsi Responden Berdasarkan Gender')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Berdasarkan hasil analisis komposisi gender responden, mayoritas responden adalah perempuan dengan persentase 50,0%, diikuti oleh laki-laki sebesar 47,7%, dan kategori lainnya sebesar 2,3%. Hal ini menunjukkan bahwa dataset relatif seimbang antara responden laki-laki dan perempuan, sementara responden dengan gender lain hanya sebagian kecil dari total responden.

## **Distribution(Distribusi)**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Caffeine_mg'], bins=20, kde=True)
plt.title('Distribusi Asupan Kafein Responden')
plt.xlabel('Kafein (mg)')
plt.ylabel('Frekuensi')
plt.show()
```

Berdasarkan grafik tersebut terlihat bahwa sebagian besar responden mengonsumsi kafein dalam jumlah sedang yaitu di kisaran 150 sampai 300 mg per hari. Hanya sedikit responden yang memiliki asupan kafein sangat tinggi hingga di atas 500 mg sehingga bisa disimpulkan bahwa konsumsi kafein berlebih tidak terlalu dominan.

## **Relationship/Hubungan**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Coffee_Intake', 'Caffeine_mg']].corr(),
    annot=True,
    cmap='viridis',
    fmt='.2f')
plt.title('Korelasi antara Coffee Intake dan Caffeine (mg)')
plt.show()
```
