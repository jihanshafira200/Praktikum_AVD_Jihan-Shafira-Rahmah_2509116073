## **Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Outliers**

Penanganan outliers dilakukan untuk mengecek apakah ada nilai data yang terlalu jauh berbeda dari data lainnya. Hal ini dilakukan agar hasil analisis tidak terganggu oleh nilai nilai yang tidak wajar dan tetap menggambarkan kondisi data secara umum.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Berdasarkan hasil pengecekan outliers diketahui bahwa kolom smoking memiliki persentase outliers paling tinggi. Namun outliers tersebut tidak dilakukan penanganan karena nilai yang terdeteksi hanya menunjukkan perbedaan antara responden yang merokok dan tidak merokok. Kondisi ini bukan merupakan kesalahan data sehingga data tetap digunakan tanpa penanganan lebih lanjut.

# **Data Construction**

Pada bagian data construction ini, saya menambahkan kolom Caffeine_Risk yang merupakan turunan dari kolom Caffeine_mg. Kolom ini mengkategorikan risiko kafein tiap responden menjadi Safe, Moderate, dan High, sehingga memudahkan analisis hubungan konsumsi kafein dengan kesehatan dan kualitas tidur.


```python
def caffeine_risk(caffeine_mg):
    if caffeine_mg <= 200:
        return 'Safe'
    elif 201 <= caffeine_mg <= 400:
        return 'Moderate'
    else:
        return 'High'

df['Caffeine_Risk'] = df['Caffeine_mg'].apply(caffeine_risk)
```

Histogram digunakan untuk menampilkan distribusi atau penyebaran data numerik dengan mengelompokkan nilai ke dalam beberapa interval sehingga dapat terlihat pola frekuensi data, seperti nilai yang paling sering muncul atau rentang nilai yang paling dominan.


```python
plt.figure(figsize=(8,5))
sns.histplot(df['Coffee_Intake'], bins=20, kde=True, color='blue')

plt.title('Distribusi Konsumsi Kopi Responden')
plt.xlabel('Coffee Intake')
plt.ylabel('Frekuensi')

plt.show()
```

**Insight:**

Histogram tersebut menunjukkan bahwa distribusi konsumsi kopi responden paling banyak berada pada kisaran konsumsi sedang, yaitu sekitar 2 hingga 4 cangkir. Sementara itu, hanya sebagian kecil responden yang memiliki konsumsi kopi sangat tinggi maupun sangat rendah. Hal ini menunjukkan bahwa mayoritas responden memiliki pola konsumsi kopi yang relatif moderat.

**Action:**

Action yang dapat dilakukan adalah memberikan edukasi mengenai batas konsumsi kafein yang sehat, terutama bagi responden yang memiliki tingkat konsumsi kopi yang tinggi.

## **BOXPLOT**

Boxplot digunakan untuk menunjukkan distribusi dan penyebaran data dalam suatu dataset, serta membantu melihat adanya nilai yang menyimpang (outlier).


```python
stress_top = df['Stress_Level'].value_counts().head(5).index
stress_data = df[df['Stress_Level'].isin(stress_top)]

plt.figure(figsize=(12, 7))
sns.boxplot(x='Stress_Level', y='Coffee_Intake', data=stress_data, palette='viridis', order=stress_top, hue='Stress_Level', legend=False)
plt.title('Distribusi Konsumsi Kopi Berdasarkan Tingkat Stres', fontsize=16)
plt.xlabel('Stress Level', fontsize=12)
plt.ylabel('Coffee Intake', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Insight:**

Line chart tersebut menunjukkan bahwa rata-rata konsumsi kopi cenderung menurun seiring dengan meningkatnya jam tidur. Hal ini menunjukkan bahwa responden yang memiliki waktu tidur lebih sedikit cenderung mengonsumsi kopi lebih banyak. Selain itu, terdapat beberapa outlier pada data yang menunjukkan adanya nilai konsumsi kopi yang lebih tinggi dibandingkan pola umum.

**Action:**

Action yang dapat dilakukan adalah meningkatkan kesadaran mengenai pentingnya menjaga pola tidur yang cukup agar konsumsi kopi tidak berlebihan dan tetap berada pada batas yang wajar.

### **SCATTERPLOT**

Scatterplot digunakan untuk menampilkan hubungan atau korelasi antara dua variabel numerik dalam suatu dataset. Grafik ini membantu melihat pola hubungan, kecenderungan, serta penyebaran data antara kedua variabel tersebut.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Coffee_Intake', y='Sleep_Hours', hue='Stress_Level', data=df)

plt.title('Hubungan Konsumsi Kopi dan Jam Tidur Berdasarkan Tingkat Stres')
plt.xlabel('Coffee Intake')
plt.ylabel('Sleep Hours')
plt.legend(title='Stress Level')

plt.show()
```

**Insight:**
Scatterplot tersebut menunjukkan bahwa responden dengan tingkat stres tinggi cenderung memiliki jam tidur yang lebih sedikit dan konsumsi kopi yang lebih tinggi. Sebaliknya, responden dengan tingkat stres rendah umumnya memiliki jam tidur lebih panjang dan konsumsi kopi yang lebih rendah. Pola ini menunjukkan adanya hubungan antara tingkat stres, durasi tidur, dan kebiasaan konsumsi kopi.

**Action:**

Action yang dapat dilakukan adalah meningkatkan manajemen stres serta menjaga pola tidur yang sehat agar konsumsi kopi tetap berada pada tingkat yang wajar.

### **HEATMAP**

Heatmap digunakan untuk menampilkan tingkat hubungan atau korelasi antara beberapa variabel dalam dataset dengan menggunakan perbedaan warna, sehingga memudahkan dalam melihat pola hubungan antar variabel secara visual.


```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Age', 'Coffee_Intake', 'Sleep_Hours']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')

plt.title('Korelasi antara Usia, Konsumsi Kopi, dan Jam Tidur')
plt.show()
```

**Insight:**

Heatmap tersebut menunjukkan bahwa hubungan antara usia dengan konsumsi kopi maupun jam tidur hampir tidak ada, karena nilai korelasinya mendekati 0. Sementara itu, terdapat korelasi negatif lemah (-0,19) antara konsumsi kopi dan jam tidur, yang berarti semakin tinggi konsumsi kopi seseorang maka cenderung diikuti dengan semakin sedikit jam tidur. Namun, karena nilai korelasinya relatif kecil, hubungan ini tidak terlalu kuat dan masih dipengaruhi oleh faktor lain.

**Action:**

Action yang dapat dilakukan adalah melakukan analisis lebih lanjut terhadap faktor lain seperti tingkat stres atau kebiasaan hidup yang kemungkinan memiliki pengaruh lebih besar terhadap konsumsi kopi dan durasi tidur.
